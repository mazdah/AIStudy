- 온라인 광고를 통해 의사결정 시스템에서 예측 모델의 역할과 운영, 의사결정 설계
- 온라인 광고 송출 시스템은 예측과 의사결정 사이에 충분한 시간이 없어 사람이 개입할 수 없으며 하루에 수없이 많이 일어난다.
- 이로 인해 의사결정의 근거가 되는 예측과 예측에 기반한 의사결정이라는 두 가지 기능이 필요하다.

# 12.1 온라인 광고 비지니스

- 온라인 광고의 이해관계자
    - **미디어(매체)**: 사람의 눈에 보이는 웹 사이트나 모바일 애플리케이션
    - **광고 영역**: 미디어에서 광고를 표시하는 영역
    - **광고주**: 제품이나 서비스를 판매하고자 사용자의 주의를 끌고 싶어하는 주체(마케터 등)
    - **오디언스**: 광고를 보는 사용자
    
### 12.1.1 광고 영역 거래

- DSP: 광고주를 위한 서비스
- 광고 영역 매매 거래는 광고 거래소를 통해 이루어지며 1위 가격 경매나 봉인 입찰 1위 가격 경매 등으로 실시한다.
- 광고 실시간 경매 프로세스
    1. 오디언스가 미디어나 웹 페이지 열람
    2. 미디어나 웹 페이지는 광고를 요청
    3. 경매 시스템은 광고주에게 입찰 요청
    4. 광고주들의 입찰 응답
    5. 경매 시스템에서 경매 실시
    6. 미디어나 웹 페이지에에서 광고 표시
    7. 오디언스의 응답(클릭/전환)
    8. 광고주들에게 경매 결과 통지
    
### 12.1.2 DSP의 행동 정책

- 한정된 예산으로 최대 광고 효과를 얻을 수 있는 세 가지 방법
    - 광고를 표시할 때 오디언스의 응답 예측
    - 광고 영역의 시장가격 예측
    - 예산 안에서의 입찰 금액 최적화
- DSP가 구매할 수 있는 광고 영역은 매우 많기 때문에 사람이 직접 하나하나 파라미터를 조절할 수 없고 때문에 머신러닝 사용이 필요하다.
- 이번 장에서는 주어진 예산 안에서 최대 광고 효과를 얻는 것을 DSP 정책으로 함

### 12.1.3 1위 가격 경매의 특징

- 다른 입찰자가 제시한 최고 입찰 가격에 따라 최적 입찰 가격이 정해진다.
- 경매로 얻은 물품의 가치 - 지불 금액 = 잉여, 잉여를 최대로 하는 가격이 최적 입찰 가격
- 사적 가치<sup>private value</sup>: 입찰자마다 광고를 게시하고자 하는 시점과 영역이 서로 다름

### 12.1.4 경매 입찰 과정

- 내부 경매: 최종적으로 입찰한 광고를 하나 선택하는 과정. 일반적으로 입찰 금액이 가장 높은 광고를 선택

# 12.2 문제 정식화

### 12.2.1 시장가격과 승률

- 최적 입찰 금액은 다른 입찰자의 입찰 금액에 의존한다.
- 입찰 금액 $b(>0)$을 결정했을 때 승률 $w(b)$는 최고 입찰 금액 분포의 누적 분포 함수에 해당한다.
- 끊임없이 반복해서 일어난 사건은 확률분포로 다루는 것이 좋다.
- 실제 시장가격은 광고 영역 종류나 오디언스 속성에 따라 달라져 입찰 요청 $x$에 의존하므로 승률 함수는 $w(x, b)$로 정의한다.

### 12.2.2 효용

- 잉여 = 효용

> $효용 = \begin{cases}vy - b & \mbox{(경매에 승리했을 때)} \\ 0 & \mbox{(기타)} \end{cases}$  
> 오디언스의 응답 $y \in \{0, 1\}$, 오디언스에게서 얻은 긍정적인 응답을 금액으로 환산한 값 $v(> 0)$, 경매에 지불한 약정 금액 $b$

- N번의 경매에 대한 기대 효용

> $기대 효용 = \sum_{i=1}^N[vf(x_i) - b_i]w(x_i, b_i)$

- 경매에서 지불하는 총액은 광고주의 예산 $B(> 0)$ 이내여야 함

> $maximize(b_1...b_N) \sum_{i=1}^N[vf(x_i) - b_i]w(x_i, b_i)$  
> $subject to \qquad\qquad \sum_{i=1}^Nb_iw(x_i, b_i) \leqq B$

# 12.3 예측의 역할 및 구현

### 12.3.1 오디언스의 응답 예측

- 대표적인 오디언스 응답 예측
    - 광고 클릭률 예측<sup>click through rate prediction</sup>(CTR 예측)
    - 전환율 예측<sup>conversion rate prediction</sup>(CVR 예측)
- 기대 동작과 작용
    - 가치 있는 광고 영역과 오디언스를 높게 평가할 수 있다.
        - 평가는 입찰 금액에 반영되므로 가치 있는 광고 영역을 많이 구매할수록 획득 효용도 늘어난다.
    - 가치 없는 광고 영역을 낮게 평가할 수 있다.
        - 효용을 얻을 수 없을 때는 입찰을 억제해 불필요한 비용을 줄인다.
- 이진 분류기인 로지스틱 회귀, GBDT, Factorization Machine 등을 모델 구현에 많이 사용한다.
- 예측 결과는 확률이다.
- 추론 시간은 매우 한정적이다.
- 추론 속도에 영향을 주는 파라미터는 예측 정확도와 추론 속도의 균형을 고려해야 한다.

### 12.3.2 승률(시장가격) 예측

- 승률 에측에 기대하는 작용
    - 불필요하게 높은 입찰 억제
    - 승률 통제 가능
- 결정 트리를 이용해 조건에 따라 어떤 분포를 만들면 좋을지 학습.

# 12.4

### 12.4.1 피드백 루프

- 얻어진 학습 데이터는 시스템이 취한 행동에 대한 응답에 한정된다.
- 학습 데이터에 편차가 심하므로 학습 시 출현 빈도를 경향 점수로 패 편차를 보정하거나 의도적으로 미지의 행동을 선택하는 정책을 사용할 수 있다.

### 12.4.2 불균형 데이터

- 배너 광고가 클릭되는 경우는 매우 적어 클릭율 예측 학습 데이터 값은 대부분 음수이다.
- 이에 대처하기 위해 음수를 줄이는 네거티브 다운 샘플링<sup>Negative Down Sampling</sup>을 사용한다.
- 이를 통해 전처리와 훈련 계산 비용을 줄일 수 있다.
> 다운 샘플링 이전이 확률 복원  
> $Pcarib = \frac{p}{p + (1 - p)^{r-1}}$  
> $p$: 예측 모델의 출력 확률, $r-1$: 다운 샘플링률의 역수

### 12.4.3 카디널리티가 큰 범주형 변수

- 범주형 변수이면서 표현하는 값의 패턴(카디널리티<sup>cardinality</sup>)이 큰 경우(수천수만) 원-핫 인코딩 상태에서 L1 정규화를 해 특징량을 선택한다.
- 최근에는 범주형 변수를 압축한 표현을 신경망에 학습시켜 이용하는 방법을 많이 사용한다(엔터티 임베딩<sup>entity embedding</sup>).

### 12.4.4 잘린 데이터

- 잘려진 데이터를 분석할 경우 관측 지연 시간 분포에서 경향 점수를 구한 후 레코드별로 샘플 가중치를 보정하는 등의 방법을 사용한다.
