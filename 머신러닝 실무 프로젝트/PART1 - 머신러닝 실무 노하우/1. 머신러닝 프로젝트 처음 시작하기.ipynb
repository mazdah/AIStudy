# 1.1 머신러닝으 어떻게 동작하는가

 - **머신러닝** : 과거의 경험(데이터)를 기반으로 새로운 데이터에 대한 예측을 내림
 - **지도학습<sup>supervised learning</sup>** : 이미 알고 있는 데이터에 어떤 알고리즘을 적용하여 입력 데이터와 출력 데이터 사이의 관계를 파악하고 그 결과(모델)를 이용해 새로운 예측을 수행하는 방식
     - 학습 단계: 기존 데이토의 입력과 출력(정답)의 관계를 파악
     - 예측 단계: 새로운 입력 데이터에 대한 예측 수행
 
 > **모델**: 입력 데이터와 출력 데이터 사이의 드러나지 않은 관계를 수식이나 규칙을 이용하여 근사한 것
 
 - **비지도 학습<sup>unsupervised learning</sup>** : 입력 데이터만으로 데이터의 구조를 파악해 예측 모델을 만들어내는 방식

# 1.2 머신러닝 프로젝트의 과정

1. 문제를 정의한다.
2. 머신러닝을 사용하지 않는 방법은 없는지 검토한다.
3. 시스템을 설계한다.
4. 사용할 알고리즘을 선택한다.
5. 특징과 정답 데이터, 로그를 설계한다.
6. 데이터를 전처리 한다.
7. 학습을 수행하고 파라미터를 튜닝한다.
8. 시스템에 통합한다.

- 머신러닝으로 해결한 사례를 찾으면 중점적으로 살펴볼 3가지
1. 어떤 알고리즘을 사용했는가?
2. 어떤 데이터를 특징으로 사용했는가?
3. 머신러닝 부분을 어떻게 통합했는가?

- 프로젝트에서 중요한 것은 **머신러닝으로 해결 가능한 것과 그렇지 않은 것을 구분하는 능력과 이를 실제로 구현하는 일**
- 데이터 분석의 80%는 전처리
- 머신러닝 시스템 개발은 시행착오를 반복하는 과정(주로 4\~7의 과정)

### 1.2.1 문제 정의하기

- 목적은 구체적이고 실천할 수 있는 수준까지

### 1.2.2 머신러닝을 사용하지 않는 방법 검토하기

- 머신러닝 시스템은 다른 시스템에 비해 기술부채가 쉽게 누적된다.
- 머신러닝 시스템 구축의 어려운 점
    1. 확률적인 부분이 있어 자동 테스트가 어렵다.
    2. 오래 운용하면 사용자 경향이 변해 입력 경향도 바뀐다.
    3. 처리 파이프라인이 복잡해진다.
    4. 데이터 의존관계가 복잡해진다.
    5. 실험 코드 혹은 파라미터가 포함되기 쉽다.
    6. 개발 시스템과 운영 시스템 간의 언어/프레임워크가 제각각이기 쉽다.
 
 > 머신러닝은 사람이 모두 확인할 수 없는 대량의 데이터를 처리하는 데 사용되므로 생각하지 못한 예측 결과를 출력할 위험은 항상 존재한다.
 
- 어떤 비즈니스 문제에 머신러닝을 적용하면 좋은가
    - 대량의 데이터에 대해 고속으로 안정된 판단을 내려야 한다.
    - 예측 결과에 일정 수준의 오류가 용인된다.
    
- 우선 MVP<sup>minimum viable product</sup>를 먼저 만들어본다.
- 머신러닝은 문제 설정에서 가설 검증에 이르는 주기가 일반적인 시스템의 개발 주기보다 긴 편이다.
- 따라서 비록 머신러닝 적용을 전제로 시작한 프로젝트라도 머신러닝을 사용하지 않는 방향으로의 전환을 두려워해서는 안된다.

### 1.2.3 시스템 설계하기

- 설계의 중요한 핵심 두 가지
    1. 예측 결과를 어떻게 이용할 것인가?
    2. 예측 오류의 영향을 어떻게 흡수할 것인가?
        - 구체적인 목표 성능과 포기 지점을 정해두고 성능 개선에 무리하게 집중하면 안된다.
        
### 1.2.4 알고리즘 선택하기

- 데이터 특성을 모르는 경우: 군집화<sup>clustering</sup>같은 비지도 학습이나 시각화를 통한 검토
- 데이터 양에 따라: 온라인 학습<sup>online learning</sup>과 배치 학습<sup>batch learning</sup> 중 선택

### 1.2.5 특징, 정답 데이터, 로그 설계하기

- 어떤 정보를 사용할지 설계
- 특징<sup>feature</sup>는 예측 모델에 입력될 정보
- 모든 정보는 수치화 해야 한다. 범주형 변수<sup>categorical variable</sup>도 수치로 변환해야 한다.
- 도메인 전문가의 협조가 필요하다.
- 특징 선택 후 입력 데이터로 사용할 정답 데이터도 준비한다. (지도학습에서 필요한 정답 범주 레이블과 그 정답의 입력 데이터 쌍)
- 지도 학습에서는 품질이 높은 정답 레이블을 얼마나 확보하느냐가 중요
- 애플리케이션 로그로부터 정답 데이터를 만들기도 함

### 1.2.6 데이터 전처리

- **데이터 전처리** : 불필요한 정보를 발라내고, 데이터를 머신러닝에 사용할 수 있는 형태로 바꾸는 과정
- 수치화, 이상값 처리, 누락값 처리, 정규화

### 1.2.7 학습 및 파라미터 튜닝

- 머신러닝 알고리즘을 조정하는 파라미터를 시행착오를 거쳐가며 수정해보고 결과가 더 나은 값을 찾는 과정
- 사람이 직접 부여한 정답이나 규칙 기반으로 예측한 성능을 기준치로 하여 이를 뛰어넘는 것을 목표로 함
- 과적합 방지
    1. 교차검증으로 파라미터를 튜닝
    2. 규제화를 적용
    3. 학습곡선 확인

### 1.2.8 시스템에 통합하기

- 비즈니스 관점에서 보면 가설 검증을 위한 단계
- 예측 성능 모니터링에는 미리 사람이 작성한 데이터와 정답 레이블의 쌍으로 이뤄진 데이터셋을 사용
- 시스템 통합이 완료되었다고 끝나는 것이 아니라 지속해서 개선해야 함
